## Tunnuslukuja

```python
# Selvitetään kiinnostavien sarakkeiden tunnuslukuja
"""
  Väkiluku, alle 15 vuotiaiden osuus, 15-64 vuotiaiden osuus, yli 64 vuotiaiden osuus,
  Perheiden lukumäärä, Asuntokuntien määrä, Vuokra-asunnoissa suvien osuus, Vähintään toiseen asteen tutkinnon
  suorittaneet, Korkea-asteen tutkinnon suorittaneet, Työllisyysaste, Työttömien osuus, Työpaikkojen määrä,
"""
df_tunnus = df[['Väkiluku, 2021', 'Alle 15-vuotiaiden osuus väestöstä, %, 2021',
               '15-64 -vuotiaiden osuus väestöstä, %, 2021', 'Yli 64-vuotiaiden osuus väestöstä, %, 2021',
               'Perheiden lukumäärä, 2020','Asuntokuntien lukumäärä, 2021', 'Vuokra-asunnoissa asuvien asuntokuntien osuus, %, 2020',
               'Vähintään toisen asteen tutkinnon suorittaneiden osuus 15 vuotta täyttäneistä, %, 2020', 'Korkea-asteen tutkinnon suorittaneiden osuus 15 vuotta täyttäneistä, %, 2020',
               'Työllisyysaste, %, 2020', 'Työttömien osuus työvoimasta, %, 2020', 'Alueella olevien työpaikkojen lukumäärä, 2020'
               ]].describe()

tunnusluvut = ['lukumäärä', 'keskiarvo', 'keskihajonta', 'pienin',
             'alaneljännes', 'mediaani', 'yläneljännes', 'suurin']
df_tunnus.index = tunnusluvut
df_tunnus.style.format("{:.2f}")
```

```python
# Havainnollistetaan alle 15 vuotiaiden osuus väestöstä %
sns.histplot(df['Alle 15-vuotiaiden osuus väestöstä, %, 2021'])
plt.ylabel('Kunnat')
```

```python
# Selvitetään kuntien työttömyysprosenttin esiintyvyyttä
#df['Työttömien osuus työvoimasta, %, 2020'].min() = 3.7
#df['Työttömien osuus työvoimasta, %, 2020'].max() = 21.6

# Luodaan työttömyysprosenteille luokkarajat ja lisätään aineistoon "Työttömien osuus" sarakkeen
työtön = [0,5,10,15,20,25]
df['Työttömien % osuus'] = pd.cut(df['Työttömien osuus työvoimasta, %, 2020'], bins = työtön, right=False)

df
```

```python
# Luodaan frekvenssitaulukko
df1 = pd.crosstab(df['Työttömien % osuus'], columns='f')
n = df1['f'].sum()

df1.loc['Yhteensä'] = n
df1
```

```python
sns.histplot(df['Työttömien osuus työvoimasta, %, 2020'], bins=työtön)
plt.ylabel('Kunnat')

#Esiintyvyys tiheydeltään suurin työttömyyysprosentti vaihtelee 10 ja 15 prosenttien välillä
```

## Hajontataulukko

```python
# Tarkastetaan korkea-asteen tutkinnon suorittaneiden ja palvelutyöpaikkojen osuuksien suhdetta ja huomataan että tutkinnon suorittaneet saattavat
# käyttää enemmän palveluita
sns.scatterplot(data=df, x='Korkea-asteen tutkinnon suorittaneiden osuus 15 vuotta täyttäneistä, %, 2020', y='Palvelujen työpaikkojen osuus, %, 2020', )
```

```python
# Tarkistetaan korrelaatiokerroin
# positiivinen lineaarinen yhteys
df['Korkea-asteen tutkinnon suorittaneiden osuus 15 vuotta täyttäneistä, %, 2020'].corr(df['Palvelujen työpaikkojen osuus, %, 2020'])
```

```python
# Tästä hajontataulukosta voidaan päätellä että 15-64 vuotiaat suosivat tiheemmin asuttuja alueita
sns.scatterplot(data=df, x='Taajama-aste, %, 2020', y='15-64 -vuotiaiden osuus väestöstä, %, 2021' )
```

```python
# Tarkistetaan korrelaatiokerroin
# positiivinen lineaarinen yhteys
df['Taajama-aste, %, 2020'].corr(df['15-64 -vuotiaiden osuus väestöstä, %, 2021'])
```

```python
# 64- vuotiaat suosivat vähemmän tiheäeti asuttuja alueita
sns.scatterplot(data=df, x='Yli 64-vuotiaiden osuus väestöstä, %, 2021', y='Taajama-aste, %, 2020' )
```
